```python
#import packages and libraries section 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np 
from  sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#load data set into data frame 
df = pd.read_csv('placement.csv')
df.head()
```

```python
#chech that null values are present or not if present then fill with its mean 
df.isnull().sum()
```

```python
#to train its model we have to predict its linearity by plot it 
plt.figure(figsize=(8,6))
sns.scatterplot(df,x=df['cgpa'],y=df['package'])
sns.regplot(x='cgpa', y='package', data=df, scatter=False, color='red')
plt.title('CGPA vs Package - Checking Linearity')
plt.show()
##we can see that the data is linear 
```

```python
#check its dimension , if it is one then pass in double dimension . because linerregression model expects 2d array 
print(df.cgpa.ndim)
x= df[['cgpa']]
y = df[['package']]
```

```python

best_score = -1
best_state = None
for i in range(1,500):
    x_train ,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=i)
    model =LinearRegression()
    model.fit(x_train, y_train)
    score=model.score(x_test,y_test)*100

    if score>best_score:
        best_score=score
        best_state = i

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
print(" Best Random State: ",best_state)
print(" Best R² Score:",best_score )
```

```python
#check its predict values
y_pridct = model.predict(df[['cgpa']])
y_pridct
```

```python
#now plot for its original value and preidct values 
sns.scatterplot(df,x=df['cgpa'],y=df['package'])
sns.regplot(x='cgpa', y=y_pridct, data=df, scatter=False, color='red')
plt.legend(['Origianl',['Predicted']])
plt.show()
```
